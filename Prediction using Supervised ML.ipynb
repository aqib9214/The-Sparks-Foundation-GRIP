```python
# In this regression task we will predict the percentage of marks that a student is
# expected to score based upon the number of hours they studied.

# This is a simple linear regression task as it involves just two variables.
```

# Step 1: Import the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# Step 2: Load the dataset

```python
# Dataset link: http://bit.ly/w-data
# Created the DataFrame using the dataset found using the link.

data = {
    "Hours": [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8],
    "Scores": [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.head(10)
```

# Step 3: Data Visualization

```python
plt.scatter(df['Hours'], df['Scores'], color='b')
plt.title('Study Hours vs. Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied
# and percentage of score.
```

```python
# Step 4: Splitting the Data
X = df[['Hours']]
y = df['Scores']

# Here we are using 80% of our dataset for training and 20% of the data for testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X.head(5)
```

```python
y.head(5)
```

# Step 5: Model Selection and Training


```python
model = LinearRegression()
model.fit(X_train, y_train)
```

# Step 6: Making Predictions


```python
y_pred = model.predict(X_test)
```

# Step 7: Evaluating the Model


```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

# Step 8: Visualization of Results


```python
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

# Visualizing Predictions


```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Predicted')
plt.title('Actual vs Predicted Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.legend()
plt.show()
```

# Step 9: Predicting the score for 9.25 hours/day


```python
hours = np.array([[9.25]])
predicted_score = model.predict(hours)
print(f"The predicted score for a student who studies 9.25 hours/day is {predicted_score[0]:.2f}")
```
